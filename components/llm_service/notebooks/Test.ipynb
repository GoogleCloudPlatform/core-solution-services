```python
soup = BeautifulSoup(html, 'html.parser')

tags_to_remove = ["script", "style", "footer", "nav", "aside", "form", "meta",
                "iframe", "header", "button", "input", "select", "textarea",
                "noscript", "img", "figure", "figcaption", "link",
                "h1", "h2", "h3"]
for element in soup(tags_to_remove):
  element.decompose()

paragraphs = []
for element in soup.find_all(["p", "ul"]):
  if element.name == "ul":
    items = []
    for li in element.find_all("li"):
      items.append(li.get_text())
    paragraphs.append(" ".join(items))
  else:
    paragraphs.append(element.get_text())

paragraphs
```
