```python
!pip install faker


import pandas as pd
import faker
from numpy.random import choice

# Create a Faker object for generating realistic data
fake = faker.Faker()

# Specialties and corresponding titles
specialties = ["Dermatology", "Radiology", "Anesthesiology", "Plastic Surgery", 
               "Orthopedic Surgery", "Cardiology", "Emergency Medicine", 
               "Psychiatry", "Oncology", "Gastroenterology"]
titles = ["Dermatologist", "Radiologist", "Anesthesiologist", "Plastic Surgeon",
          "Orthopedic Surgeon", "Cardiologist", "Emergency Physician", 
          "Psychiatrist", "Oncologist", "Gastroenterologist"]

# Function to randomly assign specialty and corresponding title
def get_specialty_title():
    spec = choice(specialties)
    index = specialties.index(spec)  
    return spec, titles[index]

# Function to generate provider names with 'Dr.', optionally either 'MD' or 'JD'
def generate_provider_name():
    first_name = fake.first_name()
    last_name = fake.last_name()
    degree = choice(['MD', 'JD'])
    return f"Dr. {first_name} {last_name}, {degree}"

# Create empty lists for data storage
names = []
addresses = []
phone_nums = []
med_specialties = []
prof_titles = []

# Generate 100 rows of data
for _ in range(100):
    names.append(generate_provider_name())
    addresses.append(fake.address())
    phone_nums.append(fake.phone_number())
    specialty, title = get_specialty_title()
    med_specialties.append(specialty)
    prof_titles.append(title)

# Create the DataFrame    
df = pd.DataFrame({'Provider Name': names, 
                   'Address': addresses,
                   'Phone Number': phone_nums,
                   'Medical Specialty': med_specialties,
                   'Profession Title': prof_titles})


# Print the DataFrame (or export to a CSV file)
print(df) 

```

```python
df.columns = (df.columns
                .str.replace(' ', '_', regex=True)
                .str.lower()
             )
df["address"] = df["address"].str.replace("\n", ", ")
df
```

```python
df.to_csv('ny_doctors.csv', index=False)  # Uncomment to save as a CSV file
```
